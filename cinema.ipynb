Read data into Python Pandas as a Dataframe

```python
cinema_data = pd.read_csv("cinema.csv")
# cinema_data = pd.DataFrame(cinema_data)
cinema_data.head()
```

c. Convert the unit of the budget and gross columns from $ to million $ first by
dividing by 1,000,000.

```python
# cinema_data['budget']
cinema_data["budget"] = cinema_data["budget"] / 1000000
# cinema_data['budget']


cinema_data["Gross"] = cinema_data["Gross"] / 1000000

```

```python
cinema_data
```

Rename the budget and gross columns to budget(Million $) and
gross(Million $).


```python
cinema_data = cinema_data.rename(columns={"budget": "budget (Millions $)", "Gross": "gross (Millions $)"})
# cinema_data_1
cinema_data
```

e. Create a new column called profit which contains the difference of the two columns:
gross(Million $) and budget(Million $). 

```python
cinema_data.columns
```

```python
cinema_data["profit"] =  cinema_data["gross (Millions $)"] - cinema_data["budget (Millions $)"] 
cinema_data.head()
```

f. Sort the data-frame using the profit column as reference. 

```python
cinema_data = cinema_data.sort_values("profit")
cinema_data.head()
```

g. Extract the top ten profiting movies in descending order and store them in a new dataframe called top10. 

```python
top10 = cinema_data[:10].sort_values(
    by ="profit",
    
    ascending=False)
```
